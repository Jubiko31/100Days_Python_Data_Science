# NumPy's ndarray

NumPy's most amazing feature is the **powerful** ndarray.


#### 1-Dimensional Arrays (Vectors)

```python
array_1d = np.array([1, 2, 3, 4.5])
```

# Generating and Manipulating ndarrays


```python
# .arange() methos creates a vector in specific range
a = np.arange(50, 69)
print(a)
```

```python
# Using Python slicing techniques
a[-3:]  # last 3
a[3:6]  # interval
a[9:0]  # except first 9
a[::2]  # every second value
```

```python
# Reverse order
np.flip(a)

# or
a[::-1]
```

```python
# Print out all the indices of the non-zero elements
b = np.array([6,0,0,9,0,10])
zerofree = np.nonzero(b)
print(zerofree)
```

# Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

#### Multiply `a1` with `b1`

```python
c1 = np.matmul(a1, b1)
print(c1.shape)
c1
```

# Manipulating Images as ndarrays


### To work with images we can use two very popular libraries - Scipy and PIL. <br /> 
#### The Scipy library contains an image of a racoon under 'miscellaneous' (misc). PIL will help us work with images.

```python
img = misc.face()
plt.imshow(img)
```

Convert the image to black and white. The values in `img` range from 0 to 255. 

#### Formula:
<img src="https://img-c.udemycdn.com/redactor/raw/2020-10-12_17-56-16-aff5999394e88abae2995c6d700a8cb1.png">

```python
sRGB_array = img / 255
```

```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
img_gray = sRGB_array @ grey_vals
img_gray = np.matmul(sRGB_array, grey_vals)
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(img_gray)
```
